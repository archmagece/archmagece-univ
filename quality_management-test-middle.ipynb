# 1. 교육기관에서 품질을 보증하기 위해서는 어떤 시스템을 갖출 수 있는지 설명하시오(15점)

## 교육기관 품질보증의 특징

### 당사자에 따라 다른 평가기준

* 학생 - 강의, 취업관련(지원, 수준)
* 산업계, 정부 - 학생의 수준, 학교의 연구 결과물

### 산출물에 따러 다른 수요자

* 강의 - 학생
* 서비스 - 학생
* 연구결과 - 사회, 기업체, 정부
* 학생 - 사회, 기업체, 정부

## 교육기관 품질보증 시스템

### 평가 기준 결정
교육기관은 공산품이나 단순서비스와 달리 다양한 당사자와 산출물이 존재해서 평가 기준도 다양하게 적용되고 적용되는 기준에 따라 다양한 결과가 나오게 된다.

기존 ISO9000이라는 제조업 및 서비스업 관련 기준이 있기 때문에 이를 교육기관에 적용하는게 더 빠르고 저렴하게 좋은 평가기준을 만들 수 있을 것이라고 생각된다.

### 용어 매핑

기존 ISO 9000에 사용되던 용어를 교육기관에 맞춰 매핑을 시킨다.

### 교육기관에 맞게 변형된 ISO 9000 적용

1. 최고경영자인 총장의 결정을 통해 ISO 900 프로젝트를 주도할 위원회를 조직한다
2. 품질시스템의 영역을 결정하고 ISO 9000 표준에서의 요구사항과 현재상황의 GAP을 파악한다
3. 광범위한 요소(학부,대학,연구프로그램)로부터 작은 요소(강의,실험,세미나)까지 품질시스템을 만든다
4. 앞에서 결정한 품질시스템 및 현재상황을 문서화한다
5. 품질시스템의 성과에 대한 측정지수를 개발한다
